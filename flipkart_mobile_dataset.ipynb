```python
# BRAND-WISE DISTINCT PRODUCT
product_brand=df.groupby('brand')['model'].nunique().sort_values(ascending=False)
```

```python
sns.set_style("whitegrid")
plt.figure()
plt.pie(
    product_brand,
    labels=product_brand.index,
    autopct='%1.1f%%',
    #startangle=90
)
plt.title("Top Brands by Distinct Mobile Models")
plt.show()
```

```python
pro_count=df['model'].value_counts().head(10)
pro_count
```

```python
# MODEL WISE PRODUCT COUNT
plt.figure(figsize=(7,5))
sns.barplot(x=pro_count.values, y=pro_count.index, color='#4aa142')
plt.title("Top 10 Mobile Models by Product Count")
plt.xticks(rotation =90)
plt.xlabel("Number of listings")
plt.ylabel("Mobile Model")
plt.show()
```

```python
color_count=df['base_color'].value_counts().head(5)
color_count
```

```python
# COLOUR WISE PRODUCT COUNT
sns.set_style("whitegrid")
plt.pie(
    color_count,
    labels=color_count.index,
    autopct='%1.1f%%',
    startangle=90       
)
plt.title("Top 5 Mobile Colours by Product Count")
plt.show()
```

```python
processor_count=df['processor'].value_counts()
processor_count
```

```python
# PROCESSOR WISE PRODUCT COUNT
def show_values(pct):
    total = sum(color_count)
    value = int(pct * total / 100)
    return f'{value}'
plt.figure()    
plt.pie(
    processor_count,
    labels=processor_count.index,
    autopct=show_values,
    #startangle=90
)
plt.title("Processor wise product count")
plt.show()
```

```python
sreensize_count=df['screen_size'].value_counts()
sreensize_count
```

```python
mean_sales=df.groupby('brand')['sales'].mean().reset_index().sort_values(by='sales',ascending=False)
total_sales=df.groupby('brand')['sales'].sum().reset_index().sort_values(by='brand',ascending=False)
```

```python
# BRAND WISE MEAN SALES AND TOTAL SALES

plt.figure(figsize=(10,4))

#Brand-wise Mean Sales
plt.subplot(1,2,1)
sns.barplot(x='brand', y='sales', data=mean_sales, color='#36e096')
plt.title('Brand-wise Mean Sales')
plt.xlabel('Brand')
plt.ylabel('Average sales')

#Brand-wise Total Sales
plt.subplot(1,2,2)
sns.barplot(x='brand', y='sales', data=total_sales,color='#e06d1b')
plt.title('Brand-wise Total Sales')
plt.xlabel('Brand')
plt.ylabel('Total sales')

plt.show()
```

```python
numeric_df=df.select_dtypes(include=['int64','float64'])
```

```python
corr=numeric_df.corr()
```

```python
# CORRELATION HEATMAP

plt.figure(figsize=(8,5))
sns.heatmap(
    corr,
    annot=True,
    cmap='rocket',
    fmt='.2f',
    linewidth=0.5
)
plt.title("CORRELATION HEATMAP")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
